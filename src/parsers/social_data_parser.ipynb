## Уровень бедности

```python
df = pd.read_excel('UROV_52.xlsx', sheet_name='по РФ и субъектам РФ 2018-2023', header=0, index_col=0)
df.head()
```

```python
regions = df.index.tolist()
years = pd.to_datetime(df.columns.astype(str), format='%Y').tolist()
print(regions)
print(years)
# values = [(region, year, value) for region in regions for year, value in zip(years, df.loc[region])]
# print (values)
```

```python
data = []
for region in regions:
    for year, value in zip(years, df.loc[region]):
        data.append({
            'Region': region,
            'Year': year.strftime('%Y-%m-%d'),
            'poverty_percent': value
        })
df_poverty = pd.DataFrame(data)
df_poverty['Region'] = df_poverty['Region'].str.strip()
df_poverty.head()
```

## Доходы населения

```python
df_3 = pd.read_excel('money.xls', sheet_name='Данные', header=[3, 4], index_col=0)
df_3 = df_3.rename_axis('Region')
# df_3.index = df_3.index.astype(str).str.replace('\t', '').astype(object)
df_3.head()
```

```python
regions_3 = df_3.index.tolist()
years_3 = pd.to_datetime(df_3.columns.astype(str), format='%Y').tolist()
print(regions_3)
print(years_3)
```

```python
def get_num_month(m: str) -> str:
    if m == 'январь':
        return '01'
    if m == 'февраль':
        return '02'
    if m == 'март':
        return '03'
    if m == 'апрель':
        return '04'
    if m == 'май':
        return '05'
    if m == 'июнь':
        return '06'
    if m == 'июль':
        return '07'
    if m == 'август':
        return '08'
    if m == 'сенябрь':
        return '09'
    if m == 'октябрь':
        return '10'
    if m == 'ноябрь':
        return '11'
    if m == 'декабрь':
        return '12'
    return m

years_months = []
for col in df_3.columns:
    year = pd.to_datetime(col[0], format='%Y')
    month = get_num_month(col[1])
    years_months.append((year, month))

print(years_months)
```

```python
data = []
for region in regions_3:
    for (year, month), value in zip(years_months, df_3.loc[region]):
        data.append({
            'Region': region,
            'Year': year.strftime('%Y'),
            'Month': month,
            'value_rub': value
        })
df_money = pd.DataFrame(data)
df_money['Region'] = df_money['Region'].str.strip()
df_money.head()
```

```python
df_money.loc[df_money['Region'] == 'Город Москва столица Российской Федерации город федерального значения', 'Region'] = 'г. Москва'
df_money.head()
```

```python
df_money.to_csv('money.csv', index=False, encoding='utf-8-sig')
```
